```python
#manually performing EDA by creating a heat map first:

import seaborn as sns
import matplotlib.pyplot as plt

corr = df_drop.corr()
fig, ax = plt.subplots(figsize = (14, 10))
 
sns.heatmap(ax = ax, \
            data = corr, \
            annot = True, \
            cmap = "coolwarm", \
            vmin = -1, vmax= 1, center = 0)
 
ax.set_title("Correlation Matrix of World Happiness")
 
plt.show()
```

Judging from this heatmap, the happiness score is strongly correlated to the Economy, followed by Health, Family, Freedom and Trust respectively. The happiness score does not seem to have a correlation to Generosity and it does not seem to correlate to the Year either, therefore I will be dismissing these factors in the following analysis. 

```python
# plotting pairwise relationships in the df_concat

sns.pairplot(df_drop)
```

With the incorporated histograms for each column, we can see that the values for the Happiness Score are normally distributed. The values in the Economy column are slightly left-skewed, whereas the values for columns like Family, Health and Freedom are visibly more left-skewed. The values of columns like Trust and Generosity are right-skewed.

```python
# looking at top 10 countries for Economy in 2019

ax = sns.barplot(x="Economy (GDP per Capita)", y="Country", data=df19.nlargest(10, "Economy (GDP per Capita)"),
            palette = sns.color_palette("mako", 10))

for label in ax.containers:
    ax.bar_label(label, fontsize=7)
plt.xlabel("Economy", fontsize=12);
plt.ylabel("Country", fontsize=12);
plt.title("Top 10 Countries with Highest Economy Score in 2019", fontsize=14)
plt.show()
```

```python
#looking at the whole df

ax = sns.barplot(x="Economy (GDP per Capita)", y="Country", data=df_drop.nlargest(10, "Economy (GDP per Capita)"),
            palette = sns.color_palette("mako", 10))

for label in ax.containers:
    ax.bar_label(label, fontsize=7)
plt.xlabel("Economy", fontsize=12);
plt.ylabel("Country", fontsize=12);
plt.title("Countries with Highest Economy Score", fontsize=14)
plt.show()
```

This bar chart shows that for the df_concat, only these 5 countries achieved the highest Economy scores.
